```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import seaborn as sns
dataset = 'scs'
dir_in = f'/home/james/projects/tsfc/code/code_11092024/results_oos/multiperiod/{dataset}/tensor_fig_oos_ret_rankptf_ver3/'

K = 10
data = jnp.load(dir_in + 'W_2005.pkl', allow_pickle=True)
assert jnp.allclose(1.0, jnp.linalg.norm(data['W'][K], axis=-2)) # each factor has norm 1 across lags
```

```python
data['W'].keys()
```

```python
for K in data['W'].keys():
    plt.figure(figsize=(9, 6))
    sns.heatmap(jnp.mean(data['W'][K], axis=0), cmap='viridis', cbar=True, xticklabels=range(1, K + 1), yticklabels=range(1, 37))
    plt.title('Heatmap of mean_W Matrix')
    plt.xlabel('Columns (25)')
    plt.ylabel('Rows (36)')
    plt.show()
```

```python
# Create a bar plot for the first column of mean_W
mean_W = jnp.mean(data['W'][K], axis=0)
plt.figure(figsize=(10, 6))
plt.bar(range(1, 37), mean_W[:, 0], color='skyblue')
plt.title('Bar Plot of mean_W[:, 0]')
plt.xlabel('Index (1 to 36)')
plt.ylabel('Values')
plt.xticks(range(1, 37))  # Set x-ticks to match the indices
plt.show()
```
